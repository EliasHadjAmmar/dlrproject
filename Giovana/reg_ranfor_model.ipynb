```python
## Split data to train/test set 
## - You perform data spliting: Here the 20% of the testing set to serve as sort of a un
## unknown data that you want to train the training set 
## - You want to evaluate whether the model that you built on the traiing set, wehther it 
## performs on the test set (in a robust manner)


from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
## Applying the model to make a prediction: 
    
## Apply the model to make a prediction of the training set

## Model 1: linear regression Algorithm to predrict
## That will allows us to evaluate the performance of the algorithm 

y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```
